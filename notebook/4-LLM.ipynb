```python
# Install groq library
import subprocess
subprocess.run(['pip', 'install', 'groq'], 
               capture_output=True)

# Import libraries
from groq import Groq
import pandas as pd
import joblib
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully!")
```

```python
# Setup Groq API
print("=" * 50)
print("SETTING UP groq API")
print("=" * 50)

# Your API key
API_KEY = ""

# Initialize groq
client = Groq(api_key=API_KEY)

print("✅ Gemini API connected successfully!")
```

```python
# Test GROQ AI with a simple question
print("=" * 50)
print("TESTING GROQ AI")
print("=" * 50)

# Send a simple message to groq
response = client.chat.completions.create(
    model= "llama-3.1-8b-instant" ,
    messages=[
        { "role":"user","content":
    " what is churn in simplw words?" }
    ]
)

print("🤖 Groq Response:")
print("-" * 50)
print(response.choices[0].message.content)
print("-" * 50)
print("\n✅ Groq test successful!")
```

```python
# Build AI Churn Analyst using Groq
print("=" * 50)
print("AI CHURN ANALYST")
print("=" * 50)

def analyze_customer(customer_data):
    # Build prompt with customer details
    prompt = f"""
    You are an expert customer retention analyst for a telecom company.
    
    Analyze this customer and provide insights:
    
    Customer Details:
    - Tenure: {customer_data['tenure']} months
    - Monthly Charges: ${customer_data['MonthlyCharges']}
    - Total Charges: ${customer_data['TotalCharges']}
    - Contract Type: {customer_data['Contract']}
    - Internet Service: {customer_data['InternetService']}
    - Tech Support: {customer_data['TechSupport']}
    - Payment Method: {customer_data['PaymentMethod']}
    - Churn Risk: {customer_data['churn_risk']}
    
    Please provide:
    1. Why this customer is at risk of churning (3 reasons)
    2. What retention offer to give this customer
    3. Priority level (High/Medium/Low)
    
    Keep response short and practical.
    """
    
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    
    return response.choices[0].message.content

print("✅ AI Analyst function created!")
```

```python
# Test AI Analyst on a real customer from our dataset
print("=" * 50)
print("TESTING AI ANALYST ON REAL CUSTOMER")
print("=" * 50)

# Load our clean dataset
df = pd.read_csv('../data/churn_clean.csv')

# Pick a high risk customer (churned = Yes)
high_risk_customer = df[df['Churn'] == 'Yes'].iloc[0]

# Prepare customer data
customer_data = {
    'tenure'        : high_risk_customer['tenure'],
    'MonthlyCharges': high_risk_customer['MonthlyCharges'],
    'TotalCharges'  : high_risk_customer['TotalCharges'],
    'Contract'      : high_risk_customer['Contract'],
    'InternetService': high_risk_customer['InternetService'],
    'TechSupport'   : high_risk_customer['TechSupport'],
    'PaymentMethod' : high_risk_customer['PaymentMethod'],
    'churn_risk'    : 'HIGH'
}

# Print customer details
print("👤 Customer Details:")
print("-" * 50)
for key, value in customer_data.items():
    print(f"  {key:20} : {value}")

print("\n🤖 AI Analysis:")
print("-" * 50)

# Get AI analysis
analysis = analyze_customer(customer_data)
print(analysis)
print("-" * 50)
print("\n✅ AI Analysis complete!")
```

```python
# Analyze multiple high risk customers at once
print("=" * 50)
print("BATCH CUSTOMER ANALYSIS")
print("=" * 50)

# Pick top 3 churned customers
high_risk_customers = df[df['Churn'] == 'Yes'].head(3)

for idx, row in high_risk_customers.iterrows():
    print(f"\n👤 Customer ID: {idx}")
    print(f"Tenure: {row['tenure']} months | "
          f"Monthly Charges: ${row['MonthlyCharges']} | "
          f"Contract: {row['Contract']}")
    
    customer_data = {
        'tenure'         : row['tenure'],
        'MonthlyCharges' : row['MonthlyCharges'],
        'TotalCharges'   : row['TotalCharges'],
        'Contract'       : row['Contract'],
        'InternetService': row['InternetService'],
        'TechSupport'    : row['TechSupport'],
        'PaymentMethod'  : row['PaymentMethod'],
        'churn_risk'     : 'HIGH'
    }
    
    print("\n🤖 AI Recommendation:")
    print("-" * 50)
    analysis = analyze_customer(customer_data)
    print(analysis)
    print("=" * 50)

print("\n✅ Batch analysis complete!")
```
